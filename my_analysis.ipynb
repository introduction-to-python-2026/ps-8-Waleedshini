```python
# --- STEP 0: Import Libraries ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files  # Required for downloading files from Colab

# --- STEP 1: Load the Data ---
# We use a public dataset about Global CO2 Emissions
url = "https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv"
df = pd.read_csv(url)

print("1. Data Loaded Successfully.")
print(f"   Rows: {df.shape[0]}, Columns: {df.shape[1]}")

# --- STEP 2: Save the File (Locally) ---
# First, save the file to the Colab virtual machine
filename = "global_co2_data.csv"
df.to_csv(filename, index=False)

# Trigger the browser to download the file to your computer
print(f"2. Downloading '{filename}' to your computer...")
files.download(filename)

# --- STEP 3: Process the Data ---
# We will filter the data to look only at the 'World' aggregate,
# and remove years where CO2 data is missing (NaN).
processed_df = df[df['country'] == 'World'].dropna(subset=['co2'])

print("3. Data Processed.")
print("   Showing the last 5 years of processed data:")
display(processed_df[['country', 'year', 'co2']].tail())

# --- STEP 4: Visualize ---
# Create a line chart showing CO2 emissions over time
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

sns.lineplot(data=processed_df, x='year', y='co2', linewidth=2.5, color='teal')

plt.title('Global CO2 Emissions Trend (World)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('CO2 Emissions (Million Tonnes)', fontsize=12)

print("4. Generating Graph...")
plt.show()
```
